## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Misal data sudah dimuat dalam DataFrame bernama df
df = pd.read_csv('/content/day.csv')

# Visualisasi boxplot untuk melihat distribusi 'cnt' berdasarkan 'weathersit'
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='weathersit', y='cnt')
plt.title('Distribusi Jumlah Sewa Sepeda Berdasarkan Kondisi Cuaca')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Jumlah Sewa Sepeda')
plt.show()
```

### Pertanyaan 2:

```python
# Misal data sudah dimuat dalam DataFrame bernama df

# Visualisasi scatter plot untuk menunjukkan hubungan suhu dan jumlah sewa sepeda
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='temp', y='cnt', hue='holiday', palette={0: 'blue', 1: 'red'})
plt.title('Hubungan Suhu dan Jumlah Sewa Sepeda Berdasarkan Hari Kerja/Hari Libur')
plt.xlabel('Suhu (Temp)')
plt.ylabel('Jumlah Sewa Sepeda')
plt.legend(title='Hari Libur (0 = Tidak, 1 = Ya)')
plt.show()
```

**Insight:**
- Visualisasi Pertanyaan ke 1 :
Visualisasi ini akan menunjukkan bagaimana jumlah sewa sepeda (cnt) tersebar berdasarkan kondisi cuaca yang berbeda. Boxplot memungkinkan kita untuk melihat median, kuartil, dan adanya outlier yang mungkin terjadi.
- Visualisasi Pertanyaan ke 2 :
Titik data akan digambar dengan warna yang berbeda berdasarkan kolom holiday (hari libur), dengan warna biru untuk hari kerja dan merah untuk hari libur.
